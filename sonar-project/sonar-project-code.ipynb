# Preparing data for model

```python
#Seperating Input and Target Variables
x = df.drop(columns=['Target'])
y = df['Target']
print(x.shape)
print(y.shape)
```

```python
#Treating skewness with the help of power_transform
from sklearn.preprocessing import power_transform
x = power_transform(x, method='yeo-johnson')
```

```python
#Scaling data for model using StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_x = sc.fit_transform(x)
```

# Finding Best Model

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score
```

# 1. LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

#finding best random_state
best_rState = 0
best_aScore = 0
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(scaled_x,y,test_size=0.25,random_state=i)
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    predict_y = lr.predict(x_test)
    lr_aScore = accuracy_score(y_test,predict_y)
        
    if lr_aScore>best_aScore:
        best_aScore = lr_aScore
        best_rState = i
                
print(f"Best random_state: {best_rState} with best accuracy_score: {round(best_aScore*100,2)}%")
```

```python
#Spliting the train test data with random_state = 63
x_train, x_test, y_train, y_test = train_test_split(scaled_x,y,test_size=0.25,random_state=63)
```

```python
#Hypertuning Parameters
parameters = {
    "solver": ['newton-cg','lbfgs','sag','saga'],
    "penalty": ['l2'],
    "tol": [0.0001,0.001,0.01],
    "fit_intercept": [True,False],
    "max_iter": [1000]
}
#finding best CV
best_cv = 0
best_cvScore = 0
for i in range(2,20):
    gscv_lr = GridSearchCV(LogisticRegression(),parameters)
    gscv_lr.fit(x_train,y_train)
    temp_cvScore = cross_val_score(gscv_lr.best_estimator_,x_train,y_train,cv=i).mean()
    
    if temp_cvScore>best_cvScore:
        best_cvScore = temp_cvScore
        best_cv = i
        
print(f"Best CV: {best_cv} with best cross_val_score: {best_cvScore}")
```

```python
#Building Final LogisticRegression model with hypertuned parameters
gscv_Lr = GridSearchCV(LogisticRegression(),parameters,cv=19)
gscv_Lr.fit(x_train,y_train)
print(gscv_Lr.best_params_)
```

```python
#Checking model performance
gscv_Lr_predict_y = gscv_Lr.best_estimator_.predict(x_test)

gscv_Lr_aScore = accuracy_score(y_test,gscv_Lr_predict_y)
gscv_Lr_cMatrix = confusion_matrix(y_test,gscv_Lr_predict_y)
gscv_Lr_cReport = classification_report(y_test,gscv_Lr_predict_y)

print(f"Accuracy Score: {round(gscv_Lr_aScore*100,2)}%")
print(f"Confusion Matrix:\n {gscv_Lr_cMatrix}")
print(f"Classification Report:\n{gscv_Lr_cReport}")
```

# 2. GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
#find best random_state for train test data
best_rState = 0
best_aScore = 0
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(scaled_x,y,test_size=0.25,random_state=i)
    gnb = GaussianNB()
    gnb.fit(x_train,y_train)
    predict_y = gnb.predict(x_test)
    temp_aScore = accuracy_score(y_test,predict_y)
    if temp_aScore>best_aScore:
        best_aScore = temp_aScore
        best_rState = i
        
print(f"Best random_state: {best_rState} with cross_value_score: {round(best_aScore*100,2)}%")
```
