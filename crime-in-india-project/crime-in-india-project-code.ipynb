# Mutli-Variate Analysis

```python
#Checking Correlation of each numerical features w.r.t other numerical features
df_corr = df_psr.corr()
rows = 2
cols = 3
fig,axes = plt.subplots(rows,cols,figsize=(rows*cols*3,rows*cols*2))
plt.subplots_adjust(hspace=2)
k=0
for i in range(rows):
    for j in range(cols):
        df_corr[df_corr.columns[k]].sort_values(ascending=False).drop(df_corr.columns[k]).plot.bar(ax=axes[i][j])
        axes[i][j].set_title(f"Correlation of {df_corr.columns[k]}")
        k = k + 1
        if k == 5:
            break;
            
plt.show()
```

# Label Encoding

```python
#Encoding categorical features using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = {}
encoded_values = {}
for i in df_psr.columns:
    if df_psr[i].dtypes == object:
        le[i] = LabelEncoder()
        encoded_values[i] = le[i].fit_transform(df_psr[i])
        print(f"\n%-15s => {le[i].classes_}"%i)
```

```python
#Replacing the encoded values with original values in new dataframe
df = df_psr.copy()
for i in encoded_values.keys():
    df[i] = encoded_values[i]
    
df.head() #Printing first 5 rows
```

```python
#Checking correlation of all the features with each other
df_all_corr = df.corr()

#Displaying correlation using heatmap
sns.heatmap(df_all_corr,annot=True,fmt='.2f')
plt.title("Correlation of Features")
plt.show()
```
