# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Checking features for unique values for possible encodings
for x in df.columns:
    if df[x].dtypes == object:
        print(f"%-30s :%12d"%(x,len(df[x].unique())))
```

```python
#Encoding features
encode_features = [
        'Name Prefix','Middle Initial','Gender','Quarter of Joining','Half of Joining','Month Name of Joining',
        'Short Month','DOW of Joining','Short DOW','Last % Hike','State','Region']
le = {}
encoded_value = {}
for i in encode_features:
    le[i] = LabelEncoder()
    encoded_value[i] = le[i].fit_transform(df[i])
    print(le[i].classes_)
```

```python
#Dropping non relevant features from dataset
drop_features = []
for i in df.columns:
    if df[i].dtypes == object:
        if len(df[i].unique()) > 50:
            drop_features.append(i)
            
print(drop_features)

df_new = df.copy() #copying df to df_new
df_new.drop(columns=drop_features,inplace=True) #Dropping features from df_new

print("Original Shape: ",df.shape)
print("New Shape: ",df_new.shape)
```

```python
#Replacing the encoded features with current features
for x in encode_features:
    df_new[x] = encoded_value[x]

df_new.head() #Printing first 5 rows of new df
```

# Preparing Data for Model

```python
#Separating Input and Target Data
x = df_wo.drop(columns=['Salary'])
y = df_wo['Salary']
print(x.shape)
print(y.shape)
```

```python
#Treating Skewness using boxcox
from sklearn.preprocessing import power_transform
x = power_transform(x)
```

```python
#Scaling the data using Standard Scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

# Finding the Best Model

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
#Defining function for best random_state
def get_best_random_state(r,model,x,y,test_size=0.25):
    best_rState = 0
    best_r2Score = 0
    for i in range(r):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=test_size,random_state=i)
        model.fit(x_train,y_train)
        predict_y = model.predict(x_test)
        temp_r2Score = r2_score(y_test,predict_y)
        if temp_r2Score>best_r2Score:
            best_r2Score = temp_r2Score
            best_rState = i
            
    return best_rState, best_r2Score
```

# 1. LinearRegression

```python
from sklearn.linear_model import LinearRegression
#Finding best random_state and r2_score
best_rState, best_r2Score = get_best_random_state(1000,LinearRegression(),x,y)
print(f"Best random_state: {best_rState} with best r2_score: {best_r2Score*100}%")
```

# 2. AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
#Finding best random_state and r2_score
best_rState, best_r2Score = get_best_random_state(1000,AdaBoostRegressor(),x,y)
print(f"Best random_state: {best_rState} with best r2_score: {best_r2Score*100}%")
```

# 3. SGDRegressor

```python
from sklearn.linear_model import SGDRegressor
#Finding best random_state and r2_score
best_rState, best_r2Score = get_best_random_state(1000,SGDRegressor(),x,y)
print(f"Best random_state: {best_rState} with best r2_score: {best_r2Score*100}%")
```

# 4. DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
#Finding best random_state and r2_score
best_rState, best_r2Score = get_best_random_state(1000,DecisionTreeRegressor(),x,y)
print(f"Best random_state: {best_rState} with best r2_score: {best_r2Score*100}%")
```
